**Ans:**
I chose Ansible for this task because it is simple to install and works agent-less, which makes configuration management easier.

1. Installing Ansible (Local Linux or Ubuntu EC2)

**Requirements:**

* Ubuntu/Debian system (can be on your laptop or an AWS EC2 instance)

* SSH access

* Python already installed (usually comes by default)

**Installation Steps:**

```python
sudo apt update
sudo apt install ansible -y
ansible --version
```

**2. Configuring the Ansible Inventory**

Ansible uses an inventory file to store the details of the machines it will manage.

Open the inventory file:

```python
sudo nano /etc/ansible/hosts
```

Then add the target servers (either your EC2 instances or even localhost if you just want to test):

```python
[webgroup]
3.111.45.120 ansible_user=ubuntu ansible_ssh_private_key_file=~/mykey.pem
3.109.88.240 ansible_user=ubuntu ansible_ssh_private_key_file=~/mykey.pem
```

**3. Creating and Running a Basic Configuration Script**

A. Sample Playbook â€“ apache_setup.yml

This playbook installs and starts Apache on the selected hosts.

```python
---
- name: Configure Web Server with Apache
  hosts: webgroup
  become: yes

  tasks:
    - name: Update package lists
      apt:
        update_cache: yes

    - name: Install Apache
      apt:
        name: apache2
        state: present

    - name: Start and enable Apache
      service:
        name: apache2
        state: started
        enabled: yes
```


**B. Execute the Playbook**

Run the playbook using:

```python
ansible-playbook apache_setup.yml
```

**4. Verifying Apache Installation**

After the playbook finishes, log in to your EC2 instance and check:

```python
curl http://localhost
```

If Apache is running, it will return the default HTML page content
